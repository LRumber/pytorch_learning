```python
# 校验训练数据
import hashlib

def generate_md5(file_path):
    hash_md5 = hashlib.md5()
    with open(file_path, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

def verify_md5(file_path, md5_file_path):
    with open(md5_file_path, 'r') as f:
        original_md5 = f.readline().split()[0]
    calculated_md5 = generate_md5(file_path)
    return calculated_md5 == original_md5, original_md5, calculated_md5

file_path = '/content/drive/MyDrive/sample_train.tar.gz'
md5_file_path = '/content/drive/MyDrive/sample_train.tar.gz.md5'

# 校验文件
is_valid, original_md5, calculated_md5 = verify_md5(file_path, md5_file_path)
if is_valid:
    print(f'文件 {file_path} 校验通过，MD5值匹配。')
else:
    print(f'文件 {file_path} 校验不通过，MD5值不匹配。')
    print(f'原始MD5值: {original_md5}')
    print(f'计算得到的MD5值: {calculated_md5}')
```

# 数据预处理

```python
import re
import os
import gc
import time
import joblib
import random
import pickle
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import itertools

random.seed(2022)
np.random.seed(2022)
sample_skeleton_train_path = '/content/drive/MyDrive/bisai_data/sample_skeleton_train.csv'
common_features_train_path = '/content/drive/MyDrive/bisai_data/common_features_train.csv'
sample_skeleton_test_path = '/content/drive/MyDrive/bisai_data/sample_skeleton_test.csv'
common_features_test_path = '/content/drive/MyDrive/bisai_data/common_features_test.csv'
save_path = "/content/drive/MyDrive/bisai_data/"
write_features_map_path = save_path + 'features_map.pkl'
write_features_path = save_path + 'all_features'
sparse_columns = ['101', '121', '122', '124', '125', '126', '127', '128', '129', '205', '206', '207', '210', '216', '508', '509', '702', '853', '301', '109_14', '110_14', '127_14', '150_14']
dense_columns = ['109_14', '110_14', '127_14', '150_14', '508', '509', '702', '853']
uses_columns = [col for col in sparse_columns] + ['D' + col for col in dense_columns]


def preprocess_data(mode='train'):
    assert mode in ['train', 'test']
    if mode == "test" and not os.path.exists(write_features_map_path):
        print("Error! Please run the train mode first!")
        return
    common_features_path = common_features_train_path if mode == "train" else common_features_test_path
    sample_skeleton_path = sample_skeleton_train_path if mode == "train" else sample_skeleton_test_path

    print(f"Start processing common_features_{mode}")
    common_feat_dict = {}
    with open(common_features_path, 'r') as fr:
        for line in tqdm(fr):
            line_list = line.strip().split(',')
            feat_strs = line_list[2]
            feat_dict = dict(zip(uses_columns, [[] for _ in range(len(uses_columns))]))
            for fstr in feat_strs.split('\x01'):
                filed, feat_val = fstr.split('\x02')
                feat, val = feat_val.split('\x03')
                if filed in sparse_columns:
                    feat_dict[filed].append(feat)
                if filed in dense_columns:
                    feat_dict['D' + filed].append(val)
            common_feat_dict[line_list[0]] = feat_dict

    print('join feats...')
    with open(f"{write_features_path}_{mode}.tmp", 'w') as fw:
        fw.write('click,purchase,' + ','.join(uses_columns) + '\n')
        with open(sample_skeleton_path, 'r') as fr:
            for line in tqdm(fr):
                line_list = line.strip().split(',')
                if line_list[1] == '0' and line_list[2] == '1':
                    continue
                feat_strs = line_list[5]
                feat_dict = dict(zip(uses_columns, [[] for _ in range(len(uses_columns))]))
                for fstr in feat_strs.split('\x01'):
                    filed, feat_val = fstr.split('\x02')
                    feat, val = feat_val.split('\x03')
                    if filed in sparse_columns:
                        feat_dict[filed].append(feat)
                    if filed in dense_columns:
                        feat_dict['D' + filed].append(val)
                for key, value in common_feat_dict[line_list[3]].items():
                  if key in feat_dict:
                    feat_dict[key] = list(set(feat_dict[key]).union(value))
                  else:
                    feat_dict[key] = value
                # 生成所有可能的特征组合
                combinations = list(itertools.product(*feat_dict.values()))
                for combination in combinations:
                    feats = line_list[1:3]
                    line = f'{feats[0]}, {feats[1]},'+','.join(map(str, combination)) + '\n'
                    fw.write(line)

def reduce_mem(df):
    starttime = time.time()
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in tqdm(df.columns):
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if pd.isnull(c_min) or pd.isnull(c_max):
                continue
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    end_mem = df.memory_usage().sum() / 1024**2
    print('-- Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction),time spend:{:2.2f} min'.format(end_mem, 100 * (start_mem - end_mem) / start_mem, (time.time() - starttime) / 60))
    gc.collect()
    return df


if __name__ == "__main__":
    preprocess_data(mode='train')
#    preprocess_data(mode='test')
```
